## Merge movies_metadata_df, credits_df, and keywords_df on their shared TMDb ID (id in each file).

```python
import pandas as pd
import numpy as np

# 1. Convert 'id' in movies_metadata_df from object to numeric.
#    We use errors='coerce' so that any invalid strings become NaN.
movies_metadata_df['id'] = pd.to_numeric(movies_metadata_df['id'], errors='coerce')

# 2. Drop rows where 'id' is NaN (since we can’t merge on missing IDs).
movies_metadata_df = movies_metadata_df.dropna(subset=['id'])

# 3. Convert 'id' to integer so it lines up with credits_df and keywords_df (which have int64).
movies_metadata_df['id'] = movies_metadata_df['id'].astype(int)

# 4. Merge movies_metadata_df with credits_df on 'id' (TMDb ID).
#    Left join ensures we keep all rows from movies_metadata_df, 
#    even if some movies don't appear in credits_df.
merged_df = movies_metadata_df.merge(credits_df, how='left', on='id')

# 5. Merge the result with keywords_df on 'id'.
merged_df = merged_df.merge(keywords_df, how='left', on='id')

# 6. Inspect the final merged DataFrame
print("Merged DataFrame shape:", merged_df.shape)
print(merged_df.info())
print(merged_df.head())
```

1. Some “id” values appear multiple times in either credits_df or keywords_df. When pandas merged one row from movies_metadata_df with multiple matching rows from the other DataFrame, i ended up duplicating that movie’s row for each match.

2. Some “id” values didn’t exist in movies_metadata_df but appeared in credits_df or keywords_df. In a left merge, they won’t add brand new rows by themselves, but if there’s any mismatch in how IDs are interpreted, the result can get tricky.

**JSON Parsing**

```python
import ast

def parse_json_column(df, column):
    def try_parse(val):
        if isinstance(val, str):
            try:
                return ast.literal_eval(val)
            except:
                return []
        elif isinstance(val, list):
            return val
        return []
    
    df[column] = df[column].apply(try_parse)

json_cols = ['genres', 'keywords', 'cast', 'crew', 'spoken_languages',
            'production_companies', 'production_countries', 'belongs_to_collection']

for col in json_cols:
    if col in movies_df.columns:
        parse_json_column(movies_df, col)
```

**Only about 19% of rows have a non-null budget and 16% have revenue. That’s a lot of missing data**

**Strategy: Missing Indicator + Impute**

A popular approach for neural nets is to add a binary flag for missingness.

This way, the neural net can learn that “budget was originally missing” if that’s a relevant pattern.

```python
movies_df['budget_missing'] = movies_df['budget'].isnull().astype(int) #This will be 1 for rows where budget was originally missing, and 0 otherwise.

median_budget = movies_df['budget'].median()  # This will Impute the median budget for missing values
movies_df['budget'].fillna(median_budget, inplace=True)   # This will fill the missing values with the median budget



# Now the model has two pieces of information for each movie’s budget:
# the budget value itself (which is the median if originally missing).
# A flag (budget_missing) indicating if the original was NaN or not.
```

```python
print("Missing budgets:", movies_df['budget'].isnull().sum())
print("Sample budgets:\n", movies_df['budget'].head(10))
```

```python
import ast

# ----- LET'S FINISH THIS:
# STEP 1: Impute 'revenue' & 'runtime' + Create Missing Indicators -----

# 1A. Revenue
movies_df['revenue_missing'] = movies_df['revenue'].isnull().astype(int)
median_revenue = movies_df['revenue'].median()
movies_df['revenue'] = movies_df['revenue'].fillna(median_revenue)

# 1B. Runtime
movies_df['runtime_missing'] = movies_df['runtime'].isnull().astype(int)
median_runtime = movies_df['runtime'].median()
movies_df['runtime'] = movies_df['runtime'].fillna(median_runtime)

# ----- STEP 2: Drop Low-Value or Mostly-Missing Columns -----
# 'homepage' (~17% coverage), 'tagline' (~45%), 'adult' & 'video' are rarely useful for a content-based recommender.
columns_to_drop = ['homepage', 'tagline', 'adult', 'video']
movies_df.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# ----- STEP 3: Keep Only "Released" Movies & Drop 'status' -----
movies_df = movies_df[movies_df['status'] == "Released"]
movies_df.drop(columns=['status'], inplace=True, errors='ignore')

# ----- STEP 4: Ensure JSON-Like Columns Are Parsed -----
# This code will parse these columns so each row is a Python list/dict (not a string).
json_cols = [
    'genres', 'keywords', 'cast', 'crew', 
    'spoken_languages', 'production_companies',
    'production_countries', 'belongs_to_collection'
]

def parse_json_column(df, col):
    def try_parse(val):
        if isinstance(val, str):
            try:
                return ast.literal_eval(val)
            except:
                return []
        elif isinstance(val, list):
            return val
        return []
    df[col] = df[col].apply(try_parse)

for col in json_cols:
    if col in movies_df.columns:
        parse_json_column(movies_df, col)

# ----- STEP 5: Final Check -----
print("After final cleaning:")
print(movies_df.info())
print(movies_df.isnull().sum().sort_values(ascending=False).head(10))

# Itty-bitty peek at a few rows
print(movies_df.head(3))
```

### Final Cleaning Steps Overview

1. **Dropped Duplicates** in `movies_metadata_df` to ensure each TMDb `id` corresponds to a single row.  
   
2. **Merged** Movie DataFrames  
   - Combined `movies_metadata`, `credits`, and `keywords` on the shared TMDb ID, ensuring each movie’s metadata, cast/crew, and keywords ended up in a single DataFrame.  
   - Resolved any cross-join duplicates (e.g., rows that appeared multiple times for the same ID) by dropping or unifying them.  
  
3. **Converted Numeric-Like Columns**  
   - `budget`, `revenue`, `popularity`, `vote_average`, `vote_count`, and `runtime` from strings (object dtype) to `float64`, turning zeros into `NaN` when appropriate. 
   
4. **Dropped or Imputed Missing Values** for Key Columns  
   - **Budget, Revenue, Runtime**: Created missingness indicator columns (`budget_missing`, `revenue_missing`, `runtime_missing`) and filled the missing values with the median for each column.  
   - Ensured that the model can see which movies had originally missing/unknown values.  
  
5. **Dropped Rows with Missing or Invalid Release Dates**  
   - Ensured that `release_date` was converted to a proper datetime, then dropped the small fraction of rows lacking valid dates.  
  
6. **Filtered Out Non-“Released” Movies**  
   - Kept only entries where `status` == “Released,” then dropped the `status` column.  
  
7. **Dropped Low-Value or Mostly-Empty Columns**  
   - Removed columns such as `homepage`, `tagline`, `adult`, and `video` that were often empty or added little relevance for our content-based recommender.  
  
8. **JSON Parsing** of List/Dict Columns  
   - Parsed columns like `genres`, `keywords`, `cast`, `crew`, etc., from JSON-like strings into actual Python lists/dicts, this provides easy extraction of genre names, keyword tokens, cast member data, and so on.  

9.  **Final Integrity Checks**  
   - Verified the DataFrame shape, types, and non-null counts with `.info()`, `.isnull().sum()`, and previewed data with `.head()` to confirm a clean, consistent dataset.
